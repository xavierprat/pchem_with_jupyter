# What are the equations to solve?

The **equations of motion** are the equations that allow us to predict the motion of a system as it evolves over time.

In Classical mechanics (the non-quantum one) there are three main formulations of such equations of motion: the newtonian, the lagrangian, and the hamiltonian. 

The one that you already know is the newtonian.

$$ F = \frac{dp}{dt} = m\frac{dv}{dt} = m\frac{d^{2}x}{dt^{2}} $$ 
and being force the derivative of the energy over position $$ F=-\frac{dE(x)}{dr}$$ 
we have
$$ -\frac{dE(x)}{dx} = m\frac{d^{2}x}{dt^{2}}$$

Which is a partial diferential equation which solutions would be x(t), that is a function that tell us the position (x) of particles that make up the system at any given time. In the case of a falling particle under constant gravity force the solution is a function x(t) that tell us where the particle is at any given time.

$$
x = x_0 + v_0 t + \frac{a}{2}t^2
$$

We won't be spending much time on the algebra. However, there are some steps and some terms that need to be emphasized. 

* In quantum mechanics, the equations of motion are given by solving the **Schrodinger equation**.  
$$ \hat{H}\Psi = E\Psi $$

This is a partial partial differential equation. The solution to this equation is a function. In fact, this type of equation has as a solution a collection of functions. This collection of functions are called **eigenfunctions**

Think of it like a string in a guitar (see here https://en.wikipedia.org/wiki/Eigenfunction#Vibrating_strings). In how many ways can a string in a guitar vibrate? It can vibrate with the fundamental note, then the first overtone(octave), the second, the third.. and so on. The sine functions describing all the vibrations would be called eigenfunctions and the energy of the string would be their corresponding eigenvalue.





2. **The wavefunction $\Psi$** is an eigenfunction of the Schrodinger equation: it gives information about the position of the system
3. **The Energy $E$** is an eigenvalue of the Schrodinger equation: notice that the energy cannot have any value ,it is quantized (unlike a particle falling or any other classical system).

4. The "$\hat{H}$" is called the the Hamiltonian. And it has the form:

$$\hat{H} = -\frac{\hbar^2}{2m}\frac{\delta^2}{\delta x^2} + V(x)$$

Where the variable "x" is the position. Because the particle in a box does not experience any potential inside the box, the overall Schrodinger equation to solve is 

$$
-\frac{\hbar^2}{2m}\frac{\delta^2\Psi}{\delta x^2} = E \Psi(x)
$$
and their solutions are 

Eigenfunctions (wavefunctions) :
$$
 \Psi(x)_n = \sqrt{\frac{2}{L}} sin\left(\frac{n \pi x}{L}\right)
$$
Eigenvalues (energies):
$$
 E_n =\frac{h^2 n^2}{8 m L^2}
$$


Few things to observe

* In the derivation of this equation one can see that the value of $n$ can only have an integer positive number: $n=1,2,3,4...$. Because of how the whole solution depends on the choice of $n$, this number is called a **quantum number**
* Notice that we have a collection of wavefunctions (eigenfunctions) that change depending on the value of "n". If we change "n" the wavefunction changes.
* The Schrodinger equation in this case is saying that if you derive twice the same function you should get the same function again multiplied by a parameter (eigenvalue) $E$

# The quantized energy of the particle in a box

* Will energy levels be equally spaced?
* Will they get closer at a higher quantum number "n"?
* Will a higher mass affect the energy of the particle?
* Will a larger box affect the energh of the particle?

$$
 E_n =\frac{h^2 n^2}{8 m L^2}
$$

h is Planck's constant h = 6.626e-34 and m is the mass of the particle. In order to avoid very tiny numbers, we will use m=1 and h=1, so the Energy won't be in Joules.


```python
import matplotlib.pyplot as plt
import numpy as np
#Constants: we modify them so we can plot easy numbers. It's just a change of units (no Joules now!)
h = 1
m = 1


def getEnergyParticle(mass,n,length):
    return (h**2 * n**2)/(8*mass*length**2)

nmax = 10 #show the first ten levels of the particle in a box
L = 1
fig, ax = plt.subplots()

for n in range(1,nmax+1):
    ax.axhline(getEnergyParticle(m,n,L))
ax.set_xlabel("L")
ax.set_ylabel("Energy")
ax.set_title("Energy for box size L="+str(L)+" and mass="+str(m));
```

# The wavefunctions of the particle in a box and their meaning

The wavefunction does not have as much physical meaning as its square. The square of the wavefunction $\Psi^2(x)$ gives you **the probability to find the particle**.

$$\Psi(x) = \sqrt{\frac{2}{L}} sin\left(\frac{n \pi x}{L}\right)$$
$$\Psi^2(x) = \left(\sqrt{\frac{2}{L}} sin\left(\frac{n \pi x}{L}\right)\right)^2$$

Notice that the wavefunction does not depend on the mass

```python
def psi(n,L,x):
    return np.sqrt(2/L)*np.sin(n*np.pi*x/L)
def psi_2(n,L,x):
    return np.square(psi(n,L,x))

#Values for L and x
L = 1
x_list = np.linspace(0,L,100) #this creates the numbers on the x axis
#choose a quantum number
n=1

#build the function
psi_2_list = []
psi_list = []
for x in x_list:
    psi_list.append(psi(n,L,x))
    psi_2_list.append(psi_2(n,L,x))
```

```python
fig = plt.figure(figsize=(15,10))

ax1 = plt.subplot(121) #121 means: the subplots will be 1 row 2 columns and this one is subplot number 1
ax2 = plt.subplot(122, sharey = ax1) #share-y means that it will share the y-axis with the ax1 plot

ax1.plot(x_list, psi_list)
ax1.set_title("Particle in a box: Wavefunction L="+str(L)+" and n="+str(n))
ax1.set_xlabel("L")
ax1.set_ylabel("Ψ",fontsize=17)
ax1.set_xlim(0,L)

ax2.plot(x_list, psi_2_list)
ax2.set_title("Particle in a box: Wavefunction Squared L="+str(L)+" and n="+str(n))
ax2.set_ylabel("Ψ^2",fontsize=17)
ax2.set_xlabel("L")
ax2.set_xlim(0,L);
```

# Particle in a box in 2D

If instead of just one dimension (x) we solve the Schrodinger equation in two dimensions (x and y), the wavefunctions have the following form

$$
\Psi_{n,m}(x,y)=\Psi_n(x) \Psi_m(y) = \frac{2}{L_x L_y}sin\left(\frac{n\pi}{L_x}x\right)sin\left(\frac{m\pi}{L_y}y\right)
$$

Notice that now there are two quantum numbers *n and m* and the box has two sides of length $L_x$ and $L_y$

The energy also depends on the two dimensions. If you compare it to the energy in one dimension notice that it looks like the total energy is the sum of the energy on each dimension

$$
E_{nm} = \frac{h^2}{8m}\frac{n^2}{L^{2}_x} + \frac{h^2}{8m}\frac{m^2}{L^{2}_y} = \frac{h^2}{8m}\left(\frac{n^2}{L^{2}_x} + \frac{m^2}{L^{2}_y} \right)
$$

An important concept here is **degeneracy** which is when more than one level has the same energy.


## Degeneracy

If we plot the different levels of the 2D particle in a box, we see that some levels have the same energy. For example, when the state (n=1,m=2) has the same energy as the state (n=2,m=1). We call these two states a degenerate state.

```python
#build a energy level graph for particle in a box
h =1
def getEnergyParticle2d(mass,nx,ny,Lx,Ly):
    return (h**2)/(8*mass)*( (nx**2)/(Lx**2) + (ny**2)/(Ly**2))

fig = plt.figure(figsize=(15,10))
ax = plt.subplot()
nmax = 5
Lx = 1
Ly = 1
m =1

for nx in range(1,nmax+1):
    for ny in range(1,nmax+1):
        y=getEnergyParticle2d(m,nx,ny,Lx,Ly)
        ax.hlines(y,xmin=0.2,xmax=0.8)
        if nx < ny:
            delta = 0.1
        else:
            delta = 0
        ax.text(1+delta, y, str(nx)+','+str(ny), ha='left', va='center')
        
ax.set_ylabel("Energy")
ax.set_title("Energy levels for the 2D box size Lx="+str(Lx)+",Ly="+str(Ly)+" and mass="+str(m));
plt.xlim(0,Lx*1.2);
```

# Particle in a box in 3D

Of course the particle in a box can be expanded to the three dimensions of space $x,y,z$. In this case there are three independent quantum numbers $n,m,l$

$$
E_{nml} = \frac{h^2}{8m}\left(\frac{n^2}{L^{2}_x} + \frac{m^2}{L^{2}_y} + \frac{l^2}{L^{2}_z} \right)
$$

and the wavefunctions are the sin product for the three dimensions

$$
\Psi_{n,m,l}(x,y,z)=\Psi_n(x) \Psi_m(y) \Psi_l(z) = \frac{8}{L_x L_y L_z}sin\left(\frac{n\pi}{L_x}x\right)sin\left(\frac{m\pi}{L_y}y \right) sin\left(\frac{l\pi}{L_z}z\right)
$$

Plotting a 3D function means that we need a 4th dimension which makes it hard to visualize. This is why
we need to come up with a cutoff or threshold that allows us to draw an isosurface where the function has that value.

```python
#source: https://github.com/wmfschneider/CHE30324/blob/master/Outline/CHE30324-outline.org
# conda install -c plotly plotly
import plotly.graph_objects as go
import numpy as np

#It has now three quantum numbers. Change these to see what it looks like
l = 2
m = 2
n = 1

xmesh=[]
ymesh=[]
zmesh=[]
# psi=np.array([])
psi=[]
points=25

for z in np.linspace(0.,1.,points):
    for y in np.linspace(0.,1.,points):
        for x in np.linspace(0.,1.,points):
            xmesh.append(x)
            ymesh.append(y)
            zmesh.append(z)
            value=np.sin(l*np.pi*x)*np.sin(m*np.pi*y)*np.sin(n*np.pi*z)
            psi.append(value)
#            np.append(value,psi)
#            print("{0:8.6f} {1:8.6f} {2:8.6f} {3:8.6f}".format(x,y,z,value))

# print(zmesh)
#print(psi)
iso = 0.1

fig= go.Figure(data=go.Isosurface(
    x=xmesh,
    y=ymesh,
    z=zmesh,
    value=psi,
    isomin=-iso,
    isomax=iso,
    surface_count=2,
    opacity=0.5,
    caps=dict(x_show=False, y_show=False,z_show=False)
))

fig.show()
```
